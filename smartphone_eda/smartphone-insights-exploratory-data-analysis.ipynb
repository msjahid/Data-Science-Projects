<u style="font-size: 17px; color: #7a4241;"><b>9. Num Cores:</b></u> The number of cores in the smartphone processor.<br>
<u style="font-size: 17px; color: #7a4241;"><b>10. Processor Speed:</b></u> The speed of the smartphone processor.<br>
<u style="font-size: 17px; color: #7a4241;"><b>11. Battery Capacity:</b></u> The capacity of the smartphone battery, usually measured in milliampere-hours (mAh).<br>

<u style="font-size: 17px; color: #7a4241;"><b>13. Fast Charging:</b></u> Details about the fast charging capability, such as the type of fast charging technology.<br>
<u style="font-size: 17px; color: #7a4241;"><b>14. RAM Capacity:</b></u> The amount of Random Access Memory (RAM) in the smartphone, usually measured in gigabytes (GB).<br>
<u style="font-size: 17px; color: #7a4241;"><b>15. Internal Memory:</b></u> The internal storage capacity of the smartphone, usually measured in gigabytes (GB) or terabytes (TB).<br>
<u style="font-size: 17px; color: #7a4241;"><b>16. Screen Size:</b></u> The size of the smartphone screen, typically measured diagonally in inches.<br>

```python
# List of columns with missing values
columns_with_missing = ['rating', 'processor_brand', 'num_cores', 'processor_speed', 
                        'battery_capacity', 'fast_charging', 'num_front_cameras', 
                        'os', 'primary_camera_front', 'extended_upto']


df[columns_with_missing].head()
```

```python
# Filter DataFrame to include only columns with missing values
df_missing = df[columns_with_missing]

# Description of the DataFrame for columns with missing values
df_missing.describe().style.background_gradient(cmap='Purples')
```

```python
# Calculate the mean of the "rating" column
rating_mean = df['rating'].mean()

# Fill missing values in the "rating" column with the calculated mean
df['rating'].fillna(rating_mean, inplace=True)

# Number of unique values to generate within the range of 40 to 50

# List of values to choose from
fast_charging_values = [44, 50]

# Generate random values to fill missing values in the "fast_charging" column
random_values = np.random.choice(fast_charging_values, size=df['fast_charging'].isnull().sum(), replace=True)

# Replace missing values in the "fast_charging" column with the generated random values
df.loc[df['fast_charging'].isnull(), 'fast_charging'] = random_values

# Calculate the mean of the "extended_upto" column
extended_upto_mean = df['extended_upto'].mean()

# Fill missing values in the "rating" column with the calculated mean
df['extended_upto'].fillna(extended_upto_mean, inplace=True)

# Calculate the mean of the "battery_capacity" column
battery_capacity_mean = df['battery_capacity'].mean()

# Fill missing values in the "battery_capacity" column with the calculated mean
df['battery_capacity'].fillna(battery_capacity_mean, inplace=True)

# Calculate the mean of the "processor_speed" column
processor_speed_mean = df['processor_speed'].mean()

# Fill missing values in the "processor_speed" column with the calculated mean
df['processor_speed'].fillna(processor_speed_mean, inplace=True)
```

```python
# Drop rows with missing values from the DataFrame
df.dropna(inplace=True)
```

```python
# Filter columns to include only integer and float columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Create figure and axes
fig, axs = plt.subplots(5, 4, figsize=(18, 14), facecolor='gray')

# Flatten axes for easier iteration
axs = axs.flatten()

# Set subplot background color
subplot_color = 'gray'

# Loop through each numeric column in your dataset
for i, col in enumerate(numeric_cols):
    sns.histplot(data=df, x=col, color='red', kde=True, ax=axs[i])
    axs[i].set_facecolor(subplot_color)  # Set subplot background color
    
    # Set plot title
    title = col.replace('_', ' ').title()  # Convert column name to title format
    axs[i].set_title(title, color='gold', fontname='Latin Modern Roman Light')
    
    # Set empty x-label and y-label
    axs[i].set_xlabel('')
    axs[i].set_ylabel('')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Generate a string of all processor brands concatenated
processor_brands_text = ' '.join(df['processor_brand'].dropna())

# Create the word cloud object with a gray background
wordcloud = WordCloud(width=800, height=400, background_color='gray').generate(processor_brands_text)

# Create a centered figure
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')

plt.show()
```

```python
# Calculate the counts of each operating system
os_counts = df['os'].value_counts()
# Calculate the counts of each fast charging availability
fast_charging_counts = df['fast_charging_available'].value_counts()

# Define subplot color
subplot_color = 'gray'

# Plotting
fig, axes = plt.subplots(2, 2, figsize=(14, 12), facecolor=subplot_color)

# Countplot for the distribution of OS
sns.countplot(data=df, x='os', ax=axes[0, 0])
axes[0, 0].set_xlabel('OS')
axes[0, 0].set_ylabel('Count')
axes[0, 0].set_facecolor(subplot_color)  # Set subplot background color

# Pie chart for the distribution of OS
axes[0, 1].pie(os_counts.values, labels=os_counts.index, autopct='%1.2f%%')
axes[0, 1].set_facecolor(subplot_color)  # Set subplot background color

# Set titles for the OS plots
axes[0, 0].set_title('Distribution of Operating Systems')
axes[0, 1].set_title('Distribution of Operating Systems')

# Countplot for the distribution of fast charging availability
sns.countplot(data=df, x='fast_charging_available', ax=axes[1, 0])
axes[1, 0].set_xlabel('Fast Charging Available')
axes[1, 0].set_ylabel('Count')
axes[1, 0].set_facecolor(subplot_color)  # Set subplot background color

# Pie chart for the distribution of fast charging availability
axes[1, 1].pie(fast_charging_counts.values, labels=fast_charging_counts.index, autopct='%1.2f%%')
axes[1, 1].set_facecolor(subplot_color)  # Set subplot background color

# Set titles for the fast charging availability plots
axes[1, 0].set_title('Distribution of Fast Charging Availability')
axes[1, 1].set_title('Distribution of Fast Charging Availability')

# Adjust layout
plt.tight_layout()

plt.show()
```

```python
# Calculate the count of each brand name
brand_name_counts = df['brand_name'].value_counts()

# Get the index of brand names sorted by count in ascending order
sorted_brand_names = brand_name_counts.index

# Set the background color
sns.set_style(rc = {'axes.facecolor': 'gray'})

# Plotting
plt.figure(figsize=(14, 6), facecolor='gray')  # Set facecolor to gray
sns.countplot(data=df, x='brand_name', order=sorted_brand_names)
plt.xlabel('Brand Name')
plt.xticks(rotation=90)
plt.show()
```

```python
categorical_columns = ['has_nfc','has_5g','has_ir_blaster',]

# Set the figure size
plt.figure(figsize=(16, 10), facecolor='gray')

# Loop through each categorical column
for i, feature in enumerate(categorical_columns, 1):
    # Calculate the value counts for the current feature
    counts = df[feature].value_counts()
    
    # Plot the pie chart
    plt.subplot(2, 3, i)
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
    plt.title(f'{feature.capitalize()} Distribution')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Filter numerical columns
numeric_cols = df.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix for numerical columns
corr_matrix = numeric_cols.corr()

# Set up the matplotlib figure
plt.figure(figsize=(20, 12), facecolor='gray')

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='tab20c', fmt=".2f", vmin = -1, vmax = 1)

# Add title
plt.title('Correlation Heat Map of different Mobile Features')

# Show plot
plt.show()
```
